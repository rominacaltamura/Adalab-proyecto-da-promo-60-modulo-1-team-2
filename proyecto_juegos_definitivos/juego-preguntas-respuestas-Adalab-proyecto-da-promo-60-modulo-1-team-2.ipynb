```python

calavera = '\U0001F480'
confeti = "\U0001F389"      
guiño = "\U0001F609"
pulgar_down ="\U0001F44E"
lengua_guiño = "\U0001F61C"
ola = "\U0001F30A"
caracola = "\U0001F41A"
planeta = "\U0001F30D"
cerebro = "\U0001F9E0"
koala = "\U0001F428"
matrioska = "\U0001FA86"
camello = "\U0001F42A"
isla = "\U0001F3DD\ufe0f"
everest = "\U0001F3D4\ufe0f"


def jugar():
    nombre = input("Ingresa tu nombre: ")
    print(f"¡Bienvenida Adalaber {nombre}! Vamos a ver cuánto sabes de Geografía {cerebro} {planeta}.")

    # Diccionario de preguntas y respuestas
    dic_geografia = {
        "1. ¿Cuál es el río más largo del mundo? \U0001F30A" : "amazonas",
        "2. ¿Cuál es la capital de Australia? \U0001F428" : "canberra",
        "3. ¿Qué país tiene más fronteras con otros países? \U0001FA86" : "rusia",
        "4. ¿En qué continente se encuentra el desierto del Sahara? \U0001F42A" : "africa",
        "5. ¿Cuál es el océano más grande del mundo? \U0001F41A" : "pacifico",
        "6. ¿Qué país tiene más islas en su territorio? \U0001F3DD\ufe0f" : "suecia",
        "7. ¿Cuál es la montaña más alta del mundo? \U0001F3D4\ufe0f" : "everest" 
        }

    preguntas_correctas = 0
    preguntas_incorrectas = 0

    for pregunta in dic_geografia:
        if preguntas_correctas >= 5 or preguntas_incorrectas >= 3:
            break

        print(pregunta)
        respuesta_user = input("Ingresa tu respuesta: ")

        if respuesta_user.lower() == dic_geografia[pregunta].lower():
            preguntas_correctas += 1
            print(f"¡Correcto! {confeti}")
        else:
            preguntas_incorrectas += 1
            print(f"Incorrecto {pulgar_down}")

        print("------------MARCADOR-----------")
        print("| Correctas:", preguntas_correctas,"               |")
        print("| Incorrectas:", preguntas_incorrectas,"             |")
        print("-------------------------------")

    if preguntas_correctas >= 5:
        print(f"¡Ganaste Adalaber! {confeti} Tienes buenos conocimientos de geografía {cerebro} {planeta}")
    else:
        print(f"{calavera} No ganaste esta vez. Pero puedes intentarlo de nuevo {calavera}")

    while True:
        jugar_otravez = input("\n¿Quieres jugar otra vez? (SI/NO) ")
        if jugar_otravez.lower() == 'si':
            jugar()
            break
        elif jugar_otravez.lower() == 'no':
            print(f'\nGracias por jugar con nosotras Adalaber, ¡HASTA PRONTO! {lengua_guiño}')
            break
        else:
            print(f"selecciona una opción correcta {lengua_guiño}")


# Iniciar el juego
jugar()
```
