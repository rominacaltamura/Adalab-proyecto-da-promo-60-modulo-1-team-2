```python
#PANDAS:
    #Se debe importar la biblioteca "random"

import random

#EMOJIS:
calavera = '\U0001F480'
confeti = "\U0001F389"      
guiño = "\U0001F609"
pulgar_down ="\U0001F44E"
lengua_guiño = "\U0001F61C"
piedra = "\U0001FAA8"   
papel = "\U0001F4C4"    
tijera = "\u2702\ufe0f"

#CARACTERÍSTICAS DEL JUEGO:
    #jugador1 es el usuario. Jugador 2 es el ordenador, que jugará contra el usuario con opciones aleatorias.  
    #El usuario (jugador1) debe elegir entre estas tres opciones: "piedra", "papel" o "tijera".
    #Jugador1 introducirá las opciones a través de un "input".

#CONDICIONES A CUMPLIR PARA GANAR:
    #Piedra: gana a tijera. | Pierde ante papel.
    #Papel: gana a piedra.  | Pierde ante tijera.
    #Tijera: gana a papel.  | Pierde ante piedra.

    #Se gana al obtener tres puntos.

lista_opciones=["piedra", "papel", "tijera"]

#CONTADORES E INDICACIONES:
    #Se llevará la cuenta de los puntos. 
    #El usuario recibirá en consola actualizaciones de los puntos después de cada jugada. 
    #Una jugada es cada vez que ambos jugadores se baten en duelo.
    #Se indicará cuándo la partida haya acabado con "FIN DE LA PARTIDA"

contadorjugador1=0
contadorjugador2=0

#DESARROLLO DEL JUEGO:
    #Se ha implementado un "bucle while" para que 
    #la partida acabe cuando uno de los dos jugadores llegue a tres puntos.

jugador2= random.choice(lista_opciones)

nombre_jugadora = input('¿Cómo te llamas, Adalaber?')

print (f'{confeti} ¡Bienvenida, {nombre_jugadora}, a nuestro juego de piedra {piedra}, papel {papel} o tijera {tijera}! {confeti}')
print (f'Seras el jugador 1, ¿Recuerdas como jugar, Adalaber? Volvamos a tu niña interior:')
print (f'{piedra} gana a {tijera}, {tijera} gana a {papel} y {papel} gana {piedra}!{guiño}')
print (f'¡Recuerda ganar 3 partidas para ser la vencedora! {confeti}')
print ('            ')
print (f'¿Fácil, no {nombre_jugadora}? ¡Comencemos!')

#CONDICIONES A CUMPLIR PARA GANAR

while contadorjugador1 <3 and contadorjugador2 <3:
    
    jugador1= input("Elige piedra, papel o tijera").lower().strip()
    
    if (jugador1 != "piedra") and (jugador1 != "papel") and (jugador1 != "tijera"):      
        print(f"La jugada '{jugador1}' no es válida. Por favor, elige una opción correcta.")
        print("-------------")
        continue

    if jugador1 == "piedra":
        
        if  jugador2 == "tijera":
            #gana jugador1
            contadorjugador1 +=1
            print(f"\n{nombre_jugadora} has elegido {piedra}:" , jugador1)
            print(f"tu rival ha elegido {tijera}:" , jugador2)
            print(f"{nombre_jugadora} has ganado esta ronda {guiño}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
            print("-------------------------------") 
        elif  jugador2 == "papel":
            #gana jugador2
            contadorjugador2 +=1
            print(f"\n{nombre_jugadora} has elegido {piedra}:" , jugador1)
            print(f"tu rival ha elegido {papel}:" , jugador2)
            print(f"tu rival ha ganado esta ronda {pulgar_down}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
            print("-------------------------------") 
        else:
            print(f"\n{nombre_jugadora} has elegido {piedra}:" , jugador1)
            print(f"tu rival ha elegido {piedra}:" , jugador2)
            print(f"Empate {lengua_guiño}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
            print("-------------------------------") 

    if jugador1 == "tijera":
        
        if  jugador2 == "papel":
            #gana jugador1
            contadorjugador1 +=1
            print(f"\n{nombre_jugadora} has elegido {tijera}:" , jugador1)
            print(f"tu rival ha elegido {papel}:" , jugador2)
            print(f"{nombre_jugadora} has ganado esta ronda {guiño}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
            print("-------------------------------") 
        elif  jugador2 == "piedra":
                #gana jugador2
                contadorjugador2 +=1
                print(f"\n{nombre_jugadora} has elegido {tijera}:" , jugador1)
                print(f"tu rival ha elegido {piedra}:" , jugador2)
                print(f"tu rival ha ganado esta ronda {pulgar_down}")
                print("------------MARCADOR-----------")
                print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")   
                print("-------------------------------")   
        else:
                print(f"\n{nombre_jugadora} has elegido {tijera}:" , jugador1)
                print(f"tu rival ha elegido {tijera}:" , jugador2)
                print(f"Empate {lengua_guiño}")
                print("------------MARCADOR-----------")
                print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
                print("-------------------------------") 
        
    if jugador1 == "papel":
        
        if  jugador2 == "piedra":
            #gana jugador1
            contadorjugador1 +=1
            print(f"\n{nombre_jugadora} has elegido {papel}:" , jugador1)
            print(f"Tu rival ha elegido {piedra}:" , jugador2)
            print(f"{nombre_jugadora} has ganado esta ronda {guiño}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")
            print("-------------------------------") 
        elif  jugador2 == "tijera":
                #gana jugador2
                contadorjugador2 +=1
                print(f"\n{nombre_jugadora} has elegido {papel}:" , jugador1)
                print(f"Tu rival ha elegido {tijera}:" , jugador2)
                print(f"Tu rival ha ganado esta ronda {pulgar_down}")
                print("------------MARCADOR-----------")
                print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |") 
                print("-------------------------------") 
        else:
            print(f"{nombre_jugadora} has elegido {papel}:" , jugador1)
            print(f"Tu rival ha elegido {papel}:" , jugador2)
            print(f"Empate {lengua_guiño}")
            print("------------MARCADOR-----------")
            print(f"| Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2} |")   
            print("-------------------------------")   

if contadorjugador1>3:
        print:(f"\nFIN DEL JUEGO {lengua_guiño}")
        print("------------------MARCADOR------------------")
        print(f"{confeti} {nombre_jugadora} Has ganado {confeti}!")
else:
    print(f"\nFIN DEL JUEGO {pulgar_down}")
    print("------------------MARCADOR------------------")
    print(f"{calavera} Tu rival te ha dado una paliza, Adalaber {calavera}")
                   
```
