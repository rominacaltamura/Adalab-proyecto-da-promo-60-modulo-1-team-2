```python
#PANDAS:
    #Se debe importar la biblioteca "random"
import random

#CARACTERÍSTICAS DEL JUEGO:
    #jugador1 es el usuario. jugador 2 es el ordenador, que jugará contra el usuario con opciones aleatorias.  
    #El usuario (jugador1) debe elegir entre estas tres opciones: "piedra", "papel" o "tijera".
    #Jugador1 introducirá las opciones a través de un "input".

lista_opciones = ["piedra", "papel", "tijera"]
jugador2 = random.choice(lista_opciones)

#CONDICIONES A CUMPLIR PARA GANAR:
    #Piedra: gana a tijera. | Pierde ante papel.
    #Papel: gana a piedra.  | Pierde ante tijera.
    #Tijera: gana a papel.  | Pierde ante piedra.

    #Se gana al obtener tres puntos.

#CONTADORES E INDICACIONES:
    #Se llevará la cuenta de los puntos. 
    #El usuario recibirá en consola actualizaciones de los puntos después de cada jugada. 
    #Una jugada es cada vez que ambos jugadores se baten en duelo.
    #Se indicará cuÁndo la partida haya acabado con "FIN DE LA PARTIDA"
     
contadorjugador1=0
contadorjugador2=0

#DESARROLLO DEL JUEGO:
    #Se ha implementado un "bucle while" para que 
    #la partida acabe cuando uno de los dos jugadores llegue a tres puntos.

while contadorjugador1 <3 and contadorjugador2 <3:

    #jugador1 debe introducir su opción.
    jugador1= (input("Elige piedra, papel o tijera").lower().strip())  #PONER LO DE LOWEEEEERR y STRIP!!!! 
    
    if (jugador1 != "piedra") and (jugador1 != "papel") and (jugador1 != "tijera"):       #PONER QUÉ PASA SI EL USUARIO PONE UNA RESPUESTA NO ACEPTADA.
        print(f"La jugada '{jugador1}' no es válida. Por favor, elige una opción correcta.")
    
    if jugador1 == "piedra":
        
        if  jugador2 == "tijera":
            #gana jugador1
            contadorjugador1 +=1
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Ha ganado el jugador 1 esta ronda")
            print("-------------")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")

        elif  jugador2 == "papel":
            #gana jugador2
            contadorjugador2 +=1
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Ha ganado el jugador 2 esta ronda")
            print("-------------")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")
        
        else:
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Empate")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")

    if jugador1 == "tijera":
        
        if  jugador2 == "papel":
            #gana jugador1
            contadorjugador1 +=1
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Ha ganado el jugador 1 esta ronda")
            print("-------------")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")

        elif  jugador2 == "piedra":
                #gana jugador2
                contadorjugador2 +=1
                print("El jugador 1 ha elegido la opción:" , jugador1)
                print("El jugador 2 ha elegido la opción:" , jugador2)
                print("Ha ganado el jugador 2 esta ronda")
                print("-------------")
                print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}") 

        else:
                print("El jugador 1 ha elegido la opción:" , jugador1)
                print("El jugador 2 ha elegido la opción:" , jugador2)
                print("Empate")
                print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")
        
    if jugador1 == "papel":
        
        if  jugador2 == "piedra":
            #gana jugador1
            contadorjugador1 +=1
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Ha ganado el jugador 1 esta ronda")
            print("-------------")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")

        elif  jugador2 == "tijera":
                #gana jugador2
            contadorjugador2 +=1
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Ha ganado el jugador 2 esta ronda")
            print("-------------")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}") 

        else:
            print("El jugador 1 ha elegido la opción:" , jugador1)
            print("El jugador 2 ha elegido la opción:" , jugador2)
            print("Empate")
            print(f"Jugador 1: {contadorjugador1} - Jugador 2: {contadorjugador2}")        

if contadorjugador1>3:
    print("---------")
    print(f"Ha ganado el jugador 1")
    print("FIN DE LA PARTIDA")
    print("---------")
           
else:
    print("---------")
    print(f"Ha ganado el jugador 2")
    print("FIN DE LA PARTIDA")
    
                   
```
